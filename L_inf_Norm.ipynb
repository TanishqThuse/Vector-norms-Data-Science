# **Feature Scaling Using L-Infinity Norm**

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm

# Heights in cm
heights = np.array([150, 165, 180, 190, 175])

# Compute L-infinity norm (Max height)
linf_norm = norm(heights,np.inf)

# Feature Scaling: Divide each height by the max height
scaled_heights = heights / linf_norm

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Plot Original Heights
axes[0].bar(range(len(heights)), heights, color='blue', alpha=0.6)
axes[0].set_xticks(range(len(heights)))
axes[0].set_xticklabels([f'P{i+1}' for i in range(len(heights))])
axes[0].set_ylabel('Height (cm)')
axes[0].set_title('Original Heights')
axes[0].grid(True, linestyle='--', alpha=0.6)

# Plot Scaled Heights (L-Infinity Norm)
axes[1].bar(range(len(scaled_heights)), scaled_heights, color='red', alpha=0.6)
axes[1].set_xticks(range(len(scaled_heights)))
axes[1].set_xticklabels([f'P{i+1}' for i in range(len(scaled_heights))])
axes[1].set_ylabel('Scaled Height')
axes[1].set_title('Scaled Heights (L-Infinity Norm)')
axes[1].grid(True, linestyle='--', alpha=0.6)

# Show plots
plt.tight_layout()
plt.show()

# Print Scaled Heights
print("Original Heights:", heights)
print("Scaled Heights (L-infinity norm):", scaled_heights)

#Why is This Useful?
#Prepares data for machine learning (ensures values are within [0,1] range).
#Prevents bias towards larger values (ensures fairness in calculations).
#Improves numerical stability (avoids large computations).
```

# L-Infinity Norm Loss **Calculation**

```python
# How Does L-Infinity Norm Measure Error?
#Instead of taking an average like MSE or MAE, the L-infinity norm picks the maximum absolute error.
# It focuses on the worst-case scenario, ensuring that the largest mistake is minimized.
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm

# Actual and Predicted values (example of house prices)
actual = np.array([200, 300, 400, 500, 600])
predicted = np.array([190, 310, 420, 550, 590])

# Compute absolute errors
errors = np.abs(actual - predicted)

# Compute different loss functions
linf_loss = np.max(errors)   # L-Infinity Norm (Max Error)
mae_loss = np.mean(errors)   # Mean Absolute Error
mse_loss = np.mean(errors**2)  # Mean Squared Error

# Print loss values
print("Absolute Errors:", errors)
print("L-Infinity Loss (Max Error):", linf_loss)
print("MAE (Mean Absolute Error):", mae_loss)
print("MSE (Mean Squared Error):", mse_loss)

# Plot the errors
plt.figure(figsize=(10, 5))

# Bar plot for each house's absolute error
plt.subplot(1, 2, 1)
plt.bar(range(1, len(errors) + 1), errors, color='orange')
plt.xlabel("House")
plt.ylabel("Absolute Error")
plt.title("Absolute Errors for Each House")

# Line plot comparing different loss values
plt.subplot(1, 2, 2)
plt.bar(["L-Infinity", "MAE", "MSE"], [linf_loss, mae_loss, mse_loss], color=['red', 'blue', 'green'])
plt.ylabel("Loss Value")
plt.title("Comparison of Loss Functions")

plt.tight_layout()
plt.show()

# L-Infinity Norm if you care about the biggest error.
# Use MAE for a simple, balanced error measurement.
# Use MSE if you want to punish large errors more.(AI models).
```

# L-Infinity Norm for Image **Comparison**

```python
 #L-Infinity Norm focuses on the worst pixel difference instead of an average.
# Used for image similarity detection, anomaly detection, and quality assessment.
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load two similar images (grayscale)
img1 = cv2.imread("img1.jpeg", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("img4.jpeg", cv2.IMREAD_GRAYSCALE)
print(img1)
# Compute absolute difference between images
diff = np.abs(img1 - img2)

# Compute L-infinity norm (maximum pixel difference)
linf_norm = np.max(diff)

# Display images and their difference
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title("Image 1")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
plt.title("Image 2")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(diff, cmap='hot')
plt.title(f"Difference (L Norm = {linf_norm})")
plt.axis("off")

plt.tight_layout()
plt.show()

print("L-Infinity Norm (Max Pixel Difference):", linf_norm)
```
